```python
import camera
```

```python
# загружаем точки из файла
points = np.loadtxt(os.path.join('images_input', 'house.p3d')).T
points = np.vstack((points, np.ones(points.shape[1])))

# настраиваем камеру
p = np.hstack((np.eye(3), np.array([[0], [0], [-10]])))
cam = camera.Camera(p)

# проекция
x = cam.project(points)

# изображаем проекцию на графике
plt.figure()
plt.plot(x[0], x[1], 'k.')
plt.show()
```

```python
r = 0.05 * np.random.rand(3)
rot = camera.rotation_matrix(r)

plt.figure()
for t in range(20):
    cam.p = np.dot(cam.p, rot)
    x = cam.project(points)
    plt.plot(x[0], x[1], 'k.', markersize=2)
plt.show()
```
