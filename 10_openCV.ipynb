```python
import cv2
import os
import matplotlib.pyplot as plt
```

```python
images = [cv2.imread(os.path.join('images_input', f'Kotlin_pan{num+1}.jpg')) for num in range(5)]
stitcher = cv2.Stitcher.create()
status, result = stitcher.stitch(images)
if status == cv2.STITCHER_OK:
    cv2.imwrite(os.path.join('images_output', 'Kotlin_panorama.jpg'), result)
    result = cv2.cvtColor(result, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(15,10))
    plt.imshow(result)
```
